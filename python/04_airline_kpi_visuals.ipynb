```python
# Import libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sqlalchemy import create_engine

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
# Connect Python to MySQL

engine = create_engine(
    "mysql+pymysql://root:@localhost/airline_analytics"
)
```

```python
# Load Airline KPI Data

query = """
SELECT *
FROM v_airline_performance
ORDER BY total_flights DESC
LIMIT 15;
"""

df_airline = pd.read_sql(query, engine)
df_airline
```

```python
# Total Flights by Airline (Bar Chart)

plt.figure()
sns.barplot(
    data=df_airline,
    x="airline_code",
    y="total_flights",
    palette="viridis"
)

plt.title("Top Airlines by Total Flights")
plt.xlabel("Airlines")
plt.ylabel("Total Flights")
plt.ticklabel_format(style='plain', axis='y')
plt.xticks(rotation=45)
plt.tight_layout()

plt.savefig("../visuals/airline_total_flights.png", dpi=300, bbox_inches="tight")
```

```python
# Weighted Average Arrival Delay by Airline

plt.figure()
sns.barplot(
    data=df_airline.sort_values("weighted_avg_arr_delay"),
    x="airline_code",
    y="weighted_avg_arr_delay",
    palette="magma"
)

plt.title("Weighted Average Arrival Delay by Airline")
plt.xlabel("Airline")
plt.ylabel("Avg Arrival Delay (minutes)")
plt.axhline(0, color='gray', linestyle='--')
plt.xticks(rotation=45)
plt.tight_layout()

plt.savefig("../visuals/airline_arrival_delay.png", dpi=300, bbox_inches="tight")
```

```python
# Cancellation Rate by Airline

plt.figure()
sns.barplot(
    data=df_airline.sort_values("cancellation_rate_pct"),
    x="airline_code",
    y="cancellation_rate_pct",
    palette="rocket"
)

plt.title("Cancellation Rate by Airline")
plt.xlabel("Airline")
plt.ylabel("Cancellation Rate (%)")
plt.xticks(rotation=45)
plt.tight_layout()

plt.savefig("../visuals/airline_cancellation_rate.png", dpi=300, bbox_inches="tight")
```
